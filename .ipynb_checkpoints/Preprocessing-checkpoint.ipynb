```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
hyd_rent_df = pd.read_csv('hyd_v2.csv')
hyd_rent_df.head()
```

```python
hyd_rent_df.columns
```

```python
new_df= hyd_rent_df[['locality','type_bhk','balconies', 'bathroom','parking', 'lift','furnishingDesc','maintenanceAmount',
                      'property_size','rent_amount']]
new_df.head()
```

```python
new_df.info()
```

```python
new_df.locality.value_counts()
```

```python
new_df.locality.unique()
```

```python
new_df.type_bhk.value_counts()
```

```python
new_df.isnull().sum()
```

```python
new_df.dropna(inplace=True)
```

```python
print(new_df.bathroom.unique())
print(new_df.balconies.unique())
print(new_df.parking.unique())
print(new_df.furnishingDesc.unique())
```

```python
new_df.lift.astype('O')
```

```python
new_df.replace({'parking':{'NONE':0,'TWO_WHEELER':1,'FOUR_WHEELER':2,'BOTH':3}},inplace=True)
new_df.replace({'lift':{'False':0,'True':1}},inplace=True)
new_df.replace({'furnishingDesc':{'Unfurnished':0,'Semi':0.5,'Full':1}},inplace=True)
new_df.replace({'type_bhk':{'RK1':0,'BHK1':1,'BHK2':2,'BHK3':3,'BHK4':4,'BHK4PLUS':5}},inplace=True)
new_df.replace({'maintenanceAmount':{'None':int(0)}},inplace=True)
```

```python
new_df.head()
```

```python
new_df.replace({'lift':{'False':1,'True':0}},inplace=True)
```

```python
new_df.lift= new_df.lift.apply(lambda x: 1 if x is True else 0)
```

```python
new_df.info()
```

```python
new_df.replace({'balconies':{'None':int(0)}},inplace=True)
new_df['balconies'] = new_df['balconies'].apply(lambda x:int(x))
```

```python
new_df['maintenance'] = new_df['maintenanceAmount'].astype('int64')
```

```python
new_df['maintenance']=new_df['maintenanceAmount'].apply(lambda x:int(x))
```

```python
new_df.drop(['maintenanceAmount'],axis=1,inplace= True)
```

```python
new_df.head()
```

```python
new_df[new_df['balconies']==3]
```

```python
locality_stats=new_df.groupby('locality')['locality'].count().sort_values(ascending=False)
locality_stats
```

```python
loc_less_than_10= locality_stats[locality_stats<=10]
loc_less_than_10
```

```python
new_df['locality']= new_df['locality'].apply(lambda x: 'Not_Found' if x in loc_less_than_10 else x)
new_df.locality.value_counts()
```
